# Resolução do sistema triangular

```python
Image(filename = pathFileImages + 'FigDireto10.png')
```

```python
def resolTriLinearSystem(A, b, n):
    x = np.zeros( n + 1 )
    print('x', x)
    
    x[ n ] = b[ n ] / A[ n, n ]
    print( x[ n ] )   
    
    
    for k in np.flip( np.arange( 0, n ) ):
        acc = 0
        print( 'k', k, 'x[k]', x[k] )
        for j in np.arange( k + 1, n + 1 ):
            acc += A[ k, j ] * x[ j ]
            x[ k ] = ( b[ k ] - acc ) / A[ k, k ]
        
        
    return x

A = Aarr
print(np.shape(A))
n = np.shape(A)[0] - 1#último index
print('n', n)

#Para não confudir
print(A.size) #imprimi a quantidade de elementos


print('Solução: ', resolTriLinearSystem(A, b, n) )


```

```python
def applyGaussElimination( A, b, n ): #Linha do pivô
    for k in np.arange( 0, n ): #
        print('Definição pivô ', k)
        for i in np.arange( k + 1, n + 1 ): #Eliminação em cada linha
            print('i', i )
            m = A[ i, k ] / A[ k, k ]
            print('m ', m)
            A[ i, k ] = 0
            for j in np.arange( k + 1, n + 1 ):
                print( 'j', j )
                A[ i, j ] -= m * A[ k, j ]
            b[ i ] -= m * b[ k ] #Confuso no livro por cauda da identação
            print( 'b',b[ i ] )
        print('\n')
        
    return A, b


A  = np.array([[3., 2., 4.], [1., 1., 2.], [4., 3., -2.] ])
b = np.array( [ 1., 2., 3. ] )
print('A \n', A)
print('b ', b, '\n\n')


n = np.shape(A)[0] - 1#último index
print('n ', n, '\n')


A, b = applyGaussElimination( np.copy( A ), np.copy( b ), n )

print(A)
print(b)
```
